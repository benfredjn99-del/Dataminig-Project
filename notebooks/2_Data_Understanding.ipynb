```python
import matplotlib.pyplot as plt

data['Accident_Severity'].value_counts().plot(kind='bar')
plt.show()
```

```python
#Analyse des relations: gravité vs vitesse 
import seaborn as sns

sns.boxplot(x='Accident_Severity', y='Speed_limit', data=data)
plt.show()
```

```python
#Distribution de la gravité 
sns.countplot(x='Accident_Severity', data=data)
plt.title("Distribution de la gravité")
plt.show()
```

```python
#Relation gravité / météo 
data = data.reset_index(drop=True)
plt.figure(figsize=(10,5))
sns.countplot(x='Weather_Conditions', hue='Accident_Severity', data=data)
plt.xticks(rotation=45)
plt.show()
```

```python
#Relation gravité / zone 
sns.countplot(x='Urban_or_Rural_Area', hue='Accident_Severity', data=data)
plt.title("Zone vs Gravité")
plt.show()
```

```python
#Analyse heure 
data['Hour'] = pd.to_datetime(data['Time'], format='%H:%M').dt.hour

sns.histplot(data['Hour'], bins=24)
plt.title("Distribution des accidents par heure")
plt.show()
```
